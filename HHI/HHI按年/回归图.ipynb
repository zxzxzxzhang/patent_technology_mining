```r
#### 安装和加载所需的包
library(readxl)

# 读取数据
df <- read_excel('聚合物及其制造工艺.xlsx')
degree =3
poly3 <- poly(df$Year, degree = degree)

# 进行三次回归
fit <- lm(df$HHI ~ poly3)
summary(fit)

# 绘制折线图
plot(df$Year, df$HHI, type='b', xlab='Year', ylab='HHI', col = '#1E90FF',ylim=c(0,1),frame.plot = FALSE,lty = 1 ,lwd =2.5, xaxt = "n")
box(bty = "L")

# 添加回归曲线和回归方程（5次多项式）
x <- seq(min(df$Year), max(df$Year), length.out = 100)
poly5 <- poly(x, degree = degree)  # 创建5次多项式
y <- predict(fit, list(poly3 = poly5))  # 使用5次多项式进行预测
lines(x, y, col = 'red',lty = 2,lwd = 1.5)

# 构建回归方程的数学表达式
equation <- bquote(
  y == .(round(fit$coefficients[1], 3)) +
    .(round(fit$coefficients[2], 3)) * x +
    .(round(fit$coefficients[3], 3)) * x^2 -
    .(abs(round(fit$coefficients[4], 3))) * x^3
)

# 在图形中显示公式
text(2015.6, 0.9, equation)
# 计算R方
r2 <- summary(fit)$r.squared
#text(2013, 0.8, paste0('R^2 = ', round(r2, 2)))
text(2013, 0.8, expression(paste("R"^2, " = 0.32")))

# 设置X轴刻度
xticks <- sort(unique(df$Year))
axis(side = 1, at = xticks, labels = xticks, cex.axis = 0.75)
title('Polymers and  its Manufacturing Process')
```

```r
#### 安装和加载所需的包
library(readxl)

# 读取数据
df <- read_excel('涂料和粘合剂.xlsx')
degree = 4
# 创建三次多项式
poly3 <- poly(df$Year, degree = degree)

# 进行三次回归
fit <- lm(df$HHI ~ poly3)
summary(fit)

# 绘制折线图
plot(df$Year, df$HHI, type='b', xlab='Year', ylab='HHI', col = '#1E90FF',frame.plot = FALSE,lty = 1 ,lwd =2.5,ylim=c(0,1.1))
box(bty = "L")

# 添加回归曲线和回归方程（5次多项式）
x <- seq(min(df$Year), max(df$Year), length.out = 100)
poly5 <- poly(x, degree = degree)  # 创建5次多项式
y <- predict(fit, list(poly3 = poly5))  # 使用5次多项式进行预测
lines(x, y, col = 'red',lty = 2,lwd = 1.5)

equation <- paste0('Y = ', round(fit$coefficients[1], 3),' ', round(fit$coefficients[2], 3), ' * x + ', 
                   round(fit$coefficients[3], 3), ' * x^2 +', round(fit$coefficients[4], 3), ' * x^3',round(fit$coefficients[5], 3), ' * x^4')

text(2015,1.1, equation)

# 计算R方
r2 <- summary(fit)$r.squared
text(2016, 1, paste0('R^2 = ', round(r2, 2)))
# 设置X轴刻度
xticks <- sort(unique(df$Year))
axis(side = 1, at = xticks, labels = xticks)
title('涂料和粘合剂')
```
