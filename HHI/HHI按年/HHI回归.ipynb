```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['figure.dpi'] = 300
```

# 聚合物及其制造工艺

```python
# 读取Excel文件
df = pd.read_excel('聚合物及其制造工艺.xlsx')
# 绘制折线图
sns.lineplot(x='Year', y='HHI', data=df, marker='o')
plt.xlabel('Year')
plt.ylabel('HHI')
# 三次函数回归分析
x = df['Year']
y = df['HHI']
coefficients = np.polyfit(x, y, 4)
regression_line = np.polyval(coefficients, x)
# 绘制回归线
plt.plot(x, regression_line, '--', color='red', label=f'Regression curve')
plt.legend()
# 在图上标注函数
equation_text = f'y = {coefficients[0]:.3f}*x^4+{coefficients[1]:.2f}*x^3+{coefficients[2]:.2f}*x^2+{coefficients[3]:.2f}*x+{coefficients[4]:.2f}'
r_squared_text = f'R^2 = {r_squared:.2f}'
plt.text(2011, 0.8, equation_text, fontsize=10)
plt.text(2011, 0.7, r_squared_text, fontsize=10)

# 自定义x轴刻度
plt.xticks(np.arange(df['Year'].min(), df['Year'].max()+1, 1))
sns.despine()
plt.title('聚合物及其制造工艺'),plt.ylim(0, 1),plt.grid(axis='x')
# 显示图表
plt.show()
```

```python
# 计算拟合优度
y_mean = np.mean(y)
ss_total = np.sum((y - y_mean) ** 2)
ss_residual = np.sum((y - regression_line) ** 2)
r_squared = 1 - (ss_residual / ss_total)
# 打印拟合优度
print(f"拟合优度（R-squared）: {r_squared}")
```
