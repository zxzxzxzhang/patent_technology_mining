```python
import pandas as pd

# 读取CSV文件
data = pd.read_excel(r'C:\Users\RIC_ZX\Desktop\papers\ZLW\TAC分类后.xlsx')

# 拆分包含多个专利权人的字段，并取|之前的内容作为专利权人
data['[标]原始申请(专利权)人'] = data['[标]原始申请(专利权)人'].str.split('|').str[0]

# 提取专利权人和IPC分类号
patent_owners = data['[标]原始申请(专利权)人']
ipc_codes = data['类别']

# 统计每个专利权人在各个IPC分类下的专利数量和专利总数
patent_counts = {}
for owner, ipc in zip(patent_owners, ipc_codes):
    owner = owner.strip()
    if '金发' in owner:
        owner = '金发科技'
    if owner not in patent_counts:
        patent_counts[owner] = {'专利总数': 0}
    ipc_truncated = ipc
    if ipc_truncated not in patent_counts[owner]:
        patent_counts[owner][ipc_truncated] = 0
    patent_counts[owner][ipc_truncated] += 1
    patent_counts[owner]['专利总数'] += 1

# 计算每个专利权人的HHI指数和在各个IPC分类下的专利个数
results = []
for owner in patent_counts:
    total_patents = patent_counts[owner]['专利总数']
    hhi = sum((count / total_patents) ** 2 for ipc, count in patent_counts[owner].items() if ipc != '专利总数')
    ipc_counts = {ipc: count for ipc, count in patent_counts[owner].items() if ipc != '专利总数'}
    results.append({'专利权人': owner, 'HHI指数': hhi, 'IPC分类计数': ipc_counts, '专利总数': total_patents})
    
sorted_results = sorted(results, key=lambda x: x['专利总数'], reverse=True)

# 输出结果到Excel文件
df = pd.DataFrame(sorted_results)
df.to_excel('patent_results.xlsx', index=False)
```

```python
import pandas as pd
from collections import Counter

# 读取CSV文件数据
data = pd.read_excel(r'C:\Users\RIC_ZX\Desktop\papers\ZLW\美国和欧盟 已筛选TAC_(poly butylene adipate-co-terephthalate) OR TAC_(PBAT).XLSX')

# 提取IPC主分类号
ipc_codes = data['IPC主分类号'].apply(lambda x: x.split('/')[0])

# 计算每个IPC主分类号的频率
ipc_counts = dict(Counter(ipc_codes))

# 计算HHI值
hhi = sum([(count / 523) ** 2 for count in ipc_counts.values()])

# 输出结果
print("总共HHI: ", hhi)
```
