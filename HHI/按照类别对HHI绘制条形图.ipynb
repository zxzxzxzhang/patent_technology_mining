```python
import matplotlib.pyplot as plt
import seaborn as sns

data = {
'Polymers and \n its Manufacturing Process':	0.0513 ,
'Film and \n its Products':	0.0547 ,
'Coatings and \nAdhesives':	0.2840 ,
'Fiber and Pulp':	0.1276 ,
'Microorganisms or \nEnzymes, \ntheir Compositions':	0.7899, 
'Medicinal, Dental or \n Cosmetic Preparations, \nCigar or Cigarette \nMouthpieces':	0.1067 ,
}
HHI_ave = 0.5005
# 创建画布和坐标轴
fig, ax = plt.subplots(figsize=(12, 8))
# 使用Seaborn库的barplot函数绘制条形图
sns.set_palette("Set2")
sns.barplot(x=list(data.keys()), y=list(data.values()), ax=ax)
# 绘制HHI_sum水平线
ax.axhline(y=HHI_ave, color='r', linestyle='--', label='HHIave')

# 添加每个条形上的数值标签
for i, v in enumerate(data.values()):
    ax.text(i, v, f'{v:.4f}', ha='center', va='bottom')
# 设置坐标轴标签和标题
ax.set_xlabel('IPC', fontsize=20)
ax.set_ylabel('HHI')
ax.set(ylim=(0, 1))
#plt.xticks(rotation=15)
# 显示图例
ax.legend()
sns.despine()
# 调整图像布局
plt.tight_layout()
# 显示图像
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取Excel文件
data = pd.read_excel('Draw.xlsx', header= 0)

# 获取x轴和y轴数据
x = data.iloc[:, 0]
y = data.iloc[:, 1]
HHI_ave = 0.2089

# 创建画布和坐标轴
fig, ax = plt.subplots(figsize=(8, 6))

# 使用Seaborn库的barplot函数绘制条形图
sns.barplot(x=x, y=y, ax=ax)

# 添加每个条形上的数值标签
for i, v in enumerate(y):
    ax.text(i, v, f'{v:.4f}', ha='center', va='bottom')

# 设置坐标轴标签和标题
ax.axhline(y=HHI_ave, color='r', linestyle='--', label='HHIave')
ax.set_xlabel('IPC')
ax.set_ylabel('HHI')
ax.set(ylim=(0, 1))
ax.legend()

# 显示图像
plt.show()
```
