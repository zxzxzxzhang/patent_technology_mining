```python
import matplotlib.pyplot as plt
import seaborn as sns

data = {
'US10723848B2':	0.00070279,
'US10239292B2':	0.000507667,
'US10767026B2':	0.000474858,
'US10590401B2':	0.000284915,
'US10124512B2':	0.000259014,
'EP3260498B1':0.000202606,
'EP3260498B1':	0.000202606,
'US8232348B2':	9.67E-05,
'EP3260496B1'	:5.07E-05,
'US10584320B2'	:4.66E-05,
'EP3260497B1'	:4.37E-05,

}
sns.set_palette("Set2")
# 创建画布和坐标轴
fig, ax = plt.subplots(figsize=(12, 8))
# 使用Seaborn库的barplot函数绘制条形图
sns.barplot(x=list(data.keys()), y=list(data.values()), ax=ax)
# 添加每个条形上的数值标签
for i, v in enumerate(data.values()):
    ax.text(i, v, f'{v:.9f}', ha='center', va='bottom')
# 设置坐标轴标签和标题
ax.set_ylabel('Betweenness Centrality',fontsize=20)
ax.set_title('Top 10 of Betweenness Centrality',fontsize=20)

# 显示图例
ax.legend()
sns.despine()
# 调整图像布局
plt.tight_layout()
# 显示图像
plt.show()
```
