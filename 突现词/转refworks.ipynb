```python
import pandas as pd

# 读取Excel文件
df = pd.read_excel('关键词抽取distil未微调.xlsx')
df
```

```python
def remove_one(text):
    text = text.replace('; ', ';')  # 只替换第一个逗号
    return text
```

```python
df['申请日'] = df['申请日'].str[:4]
#df['关键词3'] = df['关键词3'].astype(str).apply(remove_one)
df
```

```python
df['摘要'] = df['摘要'].astype(str)
df['ISSN号码'] = df['ISSN号码'].astype(str)
df = df.astype(str)
df
```

```python
from datetime import datetime

# 定义函数将数据转换为RefWorks格式
def convert_to_refworks(row):
    refworks_data = []
    refworks_data.append("RT " + "Journal Article")

    # SR: 文章来源
    refworks_data.append("SR " + str(row['文章来源']))
    
    # A1: 作者信息
    refworks_data.append("A1 " + str(row['关键词3']))
    
    # AD: 作者所属机构
    refworks_data.append("AD " + row['[标]当前申请(专利权)人'] + ";")
    
    # T1: 文章标题
    refworks_data.append("T1 " + row['标题'])
    
    # JF: 期刊名称
    refworks_data.append("JF " + row['公开(公告)号'])
    
    # YR: 发表年份
    refworks_data.append("YR " + str(row['申请日']))
    
    # IS: 期刊号码
    refworks_data.append("IS " + str(row['公开(公告)号']))
    
    # vo: 期刊号码
    refworks_data.append("vo " + str(row['vo']))
    # OP: 期刊号码
    refworks_data.append("OP " + str(row['OP']))
    
    # K1: 关键词
    refworks_data.append("K1 " + str(row['关键词3']))
    
    # AB: 摘要信息
    refworks_data.append("AB " + row['摘要'])
    
    # SN: ISSN号码
    refworks_data.append("SN " + row['ISSN号码'])
    
    # DS: 数据源
    refworks_data.append("DS " + row['数据源'])
    
    # LK: 相关链接
    refworks_data.append("LK " + row['相关链接'])
    
    return refworks_data

# 转换数据并保存为RefWorks格式的文本文件
refworks_data = []
for _, row in df.iterrows():
    refworks_data.extend(convert_to_refworks(row))
    refworks_data.append("")  
# 添加空行分隔不同文章

with open('CNKI-refworks3.txt', 'w', encoding='utf-8') as f:
    f.write('\n'.join(refworks_data))
```
