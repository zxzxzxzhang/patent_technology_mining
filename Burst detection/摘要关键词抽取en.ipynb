```python
import re
from nltk.corpus import stopwords
def filter_text(text):    
    text = text.lower()
    # 去除括号及其内部内容
    text = re.sub(r'\([^)]*\)', '', text)
    # 去除所有符号和数字
    pattern = re.compile(r'[^a-zA-Z\s]')
    text = pattern.sub(' ', text)
    words = text.split()
    # 过滤停用词、标点符号以及人称代词、感叹词、连接词和介词
    filtered_words = [word for word in words if word not in stopwords.words('english')]
    # 拼接过滤后的单词并返回
    return ' '.join(filtered_words)
df['摘要_英文'] = df['摘要_英文'].astype(str).apply(filter_text)
```

```python
import torch
from keybert import KeyBERT

# 替换为正确的模型路径
def get_top_n_words(text, top_n=3):
    kw_model = KeyBERT(r'distilbert-base-nli-mean-tokens')
    keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 1), stop_words=None, top_n=top_n)
    return keywords
```

```python
df['关键词3'] = df['摘要_英文'].apply(get_top_n_words)
df['关键词5'] = df['摘要_英文'].apply(get_top_n_words)
```

```python
import pandas as pd
import re

# 定义函数去除括号和数字
# 定义函数去除方括号、小括号和数字
def remove(text):
    text = re.sub(r'[\[\]\(\)\d\.\']', '', text)  # 去除方括号、小括号、数字、句点和单引号
    return text
```

```python
def remove_one(text):
    text = text.replace(', , ', ';')  # 只替换第一个逗号
    return text
```

```python
df['关键词3'] = df['关键词3'].astype(str).apply(remove).apply(remove_one).apply(lambda x: x[:-2])
df['关键词5'] = df['关键词5'].astype(str).apply(remove).apply(remove_one).apply(lambda x: x[:-2])
df
```
