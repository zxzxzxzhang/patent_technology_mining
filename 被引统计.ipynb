```python
import pandas as pd

df = pd.read_excel('(申请号去重后）美国和欧盟 已筛选TAC_(poly butylene adipate-co-terephthalate) OR TAC_(PBAT).xlsx')
df
```

```python
# Function to count citations
def count_citations(citations):
    if citations == "-":
        return 0
    else:
        return len(citations.split(" | "))

# Applying function to '被引用专利' column
df["被引专利数量"] = df["被引用专利"].apply(count_citations)

new_df= df[['公开(公告)号', '标题','标题(译)(简体中文)','被引专利数量','IPC主分类号']]
new_df
```

```python
new_df.to_excel('被引统计.xlsx', index=False)
```
