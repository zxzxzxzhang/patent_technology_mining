```python
import pandas as pd  
import networkx as nx  

df = pd.read_excel('data.xlsx')  
df = df[(df['摘要(全英)'].str.len() > 1) &(df['标题(全英)'].str.len() > 1)]   

df
```

```python
df.loc[:,'IPC主分类号'] = df['IPC主分类号'].str[:4]
df['IPC主分类号']
```

```python
data = df
data
```

```python
# Build the edge list
edges = []
for row in data.iterrows():
    _, patent = row
    cited_patents = patent['Cited Patents'].split(' | ')
    for cited_patent in cited_patents:
        if cited_patent != '-':
            edges.append((patent['Publication Number'], cited_patent))

from collections import Counter

# Count the occurrences of each string in the second column
string_counts = Counter(edge[1] for edge in edges)

# Keep edges where the citation count is greater than or equal to 5
filtered_edges = [edge for edge in edges if string_counts[edge[1]] >= 5]

# Create a directed graph
G = nx.DiGraph()
G.add_edges_from(filtered_edges)

# to pandas
adj_matrix = nx.to_pandas_adjacency(G, weight='weight')

adj_matrix.to_csv("patent_citation_network.csv")
```
