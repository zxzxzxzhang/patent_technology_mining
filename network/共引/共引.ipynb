```python
import pandas as pd  
import networkx as nx  

df = pd.read_excel('../../译_删除已完成_(整合版).xlsx', sheet_name='数据整合后处理')  

# 去除脏数据  
# 删除行：如果IPC主分类号、摘要_英、标题_英和中有任一字段的字符数小于或等于1，则移除这些行  
df = df[  
        (df['摘要(全英)'].str.len() > 1) &   
        (df['标题(全英)'].str.len() > 1)]   

df
```

```python
df.loc[:,'IPC主分类号'] = df['IPC主分类号'].str[:4]  #截取字符串的前四个字符
#在国际专利分类（IPC）中，前四位通常代表该专利所属的主要技术领域或类别。
df['IPC主分类号']
```

```python
data = df
data
```

```python
'''
整体流程是从专利数据中提取引用关系，统计引用次数，过滤出引用次数大于等于 5 的边，构建有向图，并生成带权邻接矩阵，最后将其保存为 CSV 文件。
'''

# 构建边列表
edges = []
for row in data.iterrows():
    _, patent = row
    cited_patents = patent['引用专利'].split(' | ')
    for cited_patent in cited_patents:
        if cited_patent != '-':
            edges.append((patent['公开(公告)号'], cited_patent))

from collections import Counter

# 统计第二列中每个字符串的数量
string_counts = Counter(edge[1] for edge in edges)

# 仅保留字符串数量大于等于5的行，
filtered_edges = [edge for edge in edges if string_counts[edge[1]] >= 5] 
#通过列表推导式，仅保留引用次数大于等于 5 的边。filtered_edges 将包含所有符合条件的边。


# 创建有向图
G = nx.DiGraph()
G.add_edges_from(filtered_edges)

# 生成带权邻接矩阵
adj_matrix = nx.to_pandas_adjacency(G, weight='weight')

# 保存邻接矩阵到CSV文件
adj_matrix.to_csv("专利引用网络带权邻接矩阵大于等于5.csv")
```
