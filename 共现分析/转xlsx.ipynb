```python
import pandas as pd
import networkx as nx

# Load the xlsx file into a DataFrame
df = pd.read_csv('[标]删除金发后的专利权人带权邻接矩阵.csv',sep=';', index_col=0)

# Extract the patent holder names
patent_holders = df.index.tolist()

# Create an empty undirected graph
G = nx.Graph()

# Add the patent holders as nodes to the graph
G.add_nodes_from(patent_holders)

# Iterate over the DataFrame
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        # If the number of collaborations is greater than 0
        if df.iloc[i, j] > 0:
            # Add an edge between the corresponding patent holders
            G.add_edge(patent_holders[i], patent_holders[j], weight=df.iloc[i, j])

# Now G is a collaboration network
```

```python
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
# 将指标结果转换为DataFrame
bc_df = pd.DataFrame(list(betweenness_centrality.items()), columns=['Node', 'Betweenness Centrality'])
cc_df = pd.DataFrame(list(closeness_centrality.items()), columns=['Node', 'Closeness Centrality'])

# 保存指标结果为CSV文件
bc_df.to_excel("betweenness_centrality_patent.xlsx", index=False)
cc_df.to_excel("closeness_centrality_patent.xlsx", index=False)
```

```python
network_centralization=nx.degree_centrality(G)
max_degree = max(network_centralization.values())
network_centralization =sum([(max_degree -degree)/(len(G)-1) for degree in network_centralization.valves()])
```
