```python
import pandas as pd

df = pd.read_excel(r'生成式人工智能（最新版）_译_国家_关键词_new.xlsx')
df
```

```python
import pandas as pd

# Convert '申请日' to datetime and extract the year
df['申请日'] = pd.to_datetime(df['申请日'])
df['Year'] = df['申请日'].dt.year

# Group by 'IPC主分类号' and 'Year', then count the patents
result = df.groupby(['Year']).size().reset_index(name='Patent Count')
result
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(result['Year'], result['Patent Count'])
plt.xlabel('Year')
plt.ylabel('Patent Count')
plt.title('Number of Patents per Year')
plt.show()
```

```python
# 划分数据为两个表：1995-2002年和2003-2010年
df1 = df[(df['申请日'] <= '2010-12-31')]
df2 = df[(df['申请日'] >= '2011-01-01') & (df['申请日'] <= '2015-12-31')]
df3 = df[(df['申请日'] >= '2016-01-01') & (df['申请日'] <= '2020-12-31')]
df4 = df[(df['申请日'] >= '2021-01-01') & (df['申请日'] <= '2022-12-31')]
df5 = df[(df['申请日'] >= '2023-01-01') & (df['申请日'] <= '2024-12-31')]
```

```python
df1.to_excel("1995-2010_data.xlsx", index=False)
df2.to_excel("2011-2015_data.xlsx", index=False)
df3.to_excel("2016-2020_data.xlsx", index=False)
df4.to_excel("2021-2022_data.xlsx", index=False)
df5.to_excel("2023-2024_data.xlsx", index=False)
```
