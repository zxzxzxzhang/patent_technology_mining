### 堆积条形图

matplotlib绘制

```python
import pandas as pd

import matplotlib.pyplot as plt


# 将"申请日"列转换为日期类型，确保数据正确解析
df['申请日'] = pd.to_datetime(df['申请日'])

# 提取年份和国家信息
df['年份'] = df['申请日'].dt.year

# 根据国家和年份进行分组统计
grouped = df.groupby(['年份', '国家']).size().unstack()

# 获取默认的颜色循环
colors = plt.cm.get_cmap('tab20', len(grouped.columns))

# 绘制堆积柱状图
plt.figure(figsize=(10, 6))
grouped.plot(kind='bar', stacked=True, colormap=colors)
plt.title('逐年申请数量的堆积柱状图')
plt.xlabel('年份')
plt.ylabel('申请数量')
plt.legend(title='国家', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

pyecharts绘制

```python
import pandas as pd
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
from pyecharts.globals import ThemeType
from pyecharts.render import make_snapshot
from snapshot_phantomjs import snapshot
# 假设 df 已经被定义并包含数据
# 将"申请日"列转换为日期类型，确保数据正确解析
df['申请日'] = pd.to_datetime(df['申请日'])

# 提取年份和国家信息
df['年份'] = df['申请日'].dt.year

# 根据国家和年份进行分组统计
grouped = df.groupby(['年份', '国家']).size().unstack().fillna(0)

# 获取年份和国家列表
years = grouped.index.tolist()
countries = grouped.columns.tolist()

# 初始化 Bar 对象
bar = Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT, bg_color="#ffffff",width='1500px',height='1000px'))

# 添加 x 轴数据（年份）
bar.add_xaxis(years)

# 添加 y 轴数据（国家对应的申请数量）
for country in countries:
    bar.add_yaxis(country, grouped[country].tolist(), stack='stack1')

# 设置全局选项
bar.set_global_opts(
    title_opts=opts.TitleOpts(),
    xaxis_opts=opts.AxisOpts(name='Year'),
    yaxis_opts=opts.AxisOpts(name='Number of patents'),
    legend_opts=opts.LegendOpts(
            pos_right="auto",  # 不固定在右侧
            pos_left="12%",  # 图例左侧距离图表左侧5%
            pos_top="10%",  # 图例顶部距离图表顶部5%
            orient="vertical",  # 垂直排列图例项
            item_width=20,  # 每个图例项的宽度
            item_gap=10,  # 图例项之间的间隔
            border_width=0
),
)

# 设置系列选项
bar.set_series_opts(
    label_opts=opts.LabelOpts(is_show=False),
    tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="shadow")
)

# 渲染图表
bar.render('逐年申请数量的堆积柱状图.html')
```

### 饼图

```python
from pyecharts import options as opts
from pyecharts.charts import Pie


# 筛选部分国家，其余为others
grouped_data = {}
others_count = 0
for country, count in countries_data.items():
    if country in ['China', 'United States', 'India', 'Israel', 'Switzerland', 'Germany', 'Canada', 'Russia', 'United Kingdom', 'Belgium', 'Australia', 'World Intellectual Property Organization', 'European Patent Organization', 'Japan', 'France', 'Netherlands Antilles', 'Singapore', 'Italy', 'South Korea']:
        grouped_data[country] = count
    else:
        others_count += count

grouped_data['Others'] = others_count

# Prepare data for the pie chart
data_pair = list(grouped_data.items())

# Create the pie chart
pie = (
    Pie()
    .add("", data_pair)
    .set_global_opts(
        #title_opts=opts.TitleOpts(title="Patent Applications by Country"),
        legend_opts=opts.LegendOpts(is_show=False, pos_right="5%")
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {d}%"))
)

# Render the chart
pie.render_notebook()
```

```python
pie.render('pie.html')
```

### 地图

```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Map
df_countries_data = pd.DataFrame(list(countries_data.items()), columns=['国家', '专利申请数'])

# 获取国家名称和数量信息
countries = df_countries_data['国家'].tolist()
quantities = df_countries_data['专利申请数'].tolist()

# 将国家名称和数量信息组合成元组列表
data_pair = [(countries[i], quantities[i]) for i in range(len(countries))]

# 找出专利申请数量前十的国家
top_10 = df_countries_data.nlargest(10, '专利申请数')['国家'].tolist()

import matplotlib.cm as cm

# 获取 Viridis 调色板的 RGB 值
num = 3
viridis_colors = cm.get_cmap('YlGnBu', num)

# 将 RGB 值转换为十六进制格式
world_map = (
    Map()
    .add("专利申请数量", data_pair, maptype="world")
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False, formatter=lambda x: f"{x.name}: {x.value}" if x.name in top_10 else "")
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Geographical distribution of patents"),
        visualmap_opts=opts.VisualMapOpts(max_=max(quantities), range_color=[f"#{int(r*255):02x}{int(g*255):02x}{int(b*255):02x}" for r, g, b, _ in viridis_colors(range(num))])
    )
)


# 显示地图
world_map.render_notebook()
```

```python
world_map.render('world_map.html')
```

```python
import pandas as pd

# Convert '申请日' to datetime and extract the year
df['申请日'] = pd.to_datetime(df['申请日'])
df['Year'] = df['申请日'].dt.year

# Group by 'IPC主分类号' and 'Year', then count the patents
result = df.groupby(['Year']).size().reset_index(name='Patent Count')
result
```

```python
# 全部国家，和下面的cell二选一
data = df

# 将日期转换为年份
data['Year'] = pd.to_datetime(data['申请日']).dt.year

# 按年份和国家计算专利数量
patent_counts = data.groupby(['Year', '国家全称']).size().reset_index(name='count')
patent_counts.to_excel('patent_counts.xlsx', index=False)
patent_counts
```

```python
# 筛选出部分需要展示的国家
countries_to_keep = [
    'China', 'United States', 'India', 'Israel', 'Switzerland',
    'Germany', 'Canada', 'Russia', 'United Kingdom', 'Belgium',
    'Australia', 'World Intellectual Property Organization',
    'European Patent Organization', 'Japan', 'France',
    'Netherlands Antilles', 'Singapore', 'Italy', 'South Korea'
]

# Replace countries not in the list with 'Others'
data['国家全称'] = data['国家全称'].apply(lambda x: x if x in countries_to_keep else 'Others')

# 按年份和国家计算专利数量
patent_counts = data.groupby(['Year', '国家全称']).size().reset_index(name='count')
patent_counts.to_excel('patent_counts_part.xlsx', index=False)
patent_counts
```

### 南丁格尔玫瑰图（如果显示不出来尝试信任当前文件）

```python
import pandas as pd
import altair as alt

# 创建数据字典
data_dict = {
    'Country': [
        'China', 'United States', 'India', 'Israel', 'Switzerland',
        'Germany', 'Canada', 'Russia', 'United Kingdom', 'Belgium',
        'Australia'
    ],
    'Value': [2001, 119, 45, 45, 41, 31, 28, 26, 23, 23, 21]
}

# 创建DataFrame
df = pd.DataFrame(data_dict)

# 创建南丁格尔玫瑰图
rose_chart = alt.Chart(df).encode(
    theta=alt.Theta("Country:N", type="nominal", sort=None, stack=True),  # 均匀分布在圆周上
    radius=alt.Radius(field="Value", type="quantitative", scale=alt.Scale(type="sqrt", zero=True, rangeMin=20)),
    color=alt.Color(field="Value", type="nominal", legend=alt.Legend(title="Country"),scale=alt.Scale(scheme='tealblues')),

    tooltip=['Country', 'Value']
).properties(
    width=550,
    height=550

)
c1 = rose_chart.mark_arc(innerRadius=200, outerRadius=200)
#c2 = rose_chart.mark_text(radiusOffset=25).encode(text="Country:N")

# 显示图表
df['Label'] = df['Value'].astype(str)+ "\n" +  df['Country'] + "\n"

# 创建带有文本标注的第二个图层
c2 = rose_chart.mark_text(radiusOffset=35, lineBreak='\n').encode(
    text=alt.Text('Label:N')  # 使用带换行符的 Label
)
c1 + c2
```
